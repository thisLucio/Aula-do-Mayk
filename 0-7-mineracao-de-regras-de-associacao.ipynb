```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
dataset
```

Criando uma nova instancia de transEncoder

```python
te = TransactionEncoder()
```

```python
te_itens = te.fit(dataset).transform(dataset)
```

```python
te_itens
```

Criando dataFrame Pandas como dados de entrada

```python
df = pd.DataFrame(te_itens, columns=te.columns_)
```

```python
df
```

Trabalhando com o algoritmo novo Apriori

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.5, use_colnames=True)
```

```python
apriori_res = apriori(df, min_support=0.5, use_colnames=True)
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(apriori_res, metric='support', min_threshold=.6)
```
